This is a super important and common question — knowing when to use parentheses () in Python can be confusing at first, but once you understand the logic, it gets much clearer.

Let’s break it down:

🧠 Rule of Thumb:
✅ Use () when you're calling a function or method (i.e., you're telling it to do something).
❌ Don’t use () when you're accessing a property or attribute (i.e., you're just getting a value).
📌 Examples You Gave:
1. pd.columns ❌ No parentheses

✅ df.columns (not pd.columns, which wouldn't work — pd is the pandas module, df is your DataFrame)
.columns is an attribute — it holds the list of column names.
You’re not calling it to do something — you’re just accessing a stored value.
df.columns        # returns list of column names
If you did df.columns(), you’d get an error: 'Index' object is not callable

2. np.sum() ✅ Use parentheses

np.sum is a function — it performs an operation.
You call it like this:
np.sum([1, 2, 3])  # returns 6
Without (), you're just referring to the function object itself:

print(np.sum)     # prints info about the function, doesn’t execute it
✅ More Real-World Examples

Code	Use ()?	Why
df.shape	❌	Just accessing a tuple, not calling anything
df.head()	✅	Calling the .head() method to return rows
df.describe()	✅	It calculates stats, so you need to call it
len(df)	✅	len is a function
df.index	❌	It's an attribute, like .columns
type(df)	✅	type() is a built-in function
df.dtypes	❌	Just checking column data types
🧪 How to Tell If It's Callable?
You can check like this:

callable(np.sum)      # True → it's a function
callable(df.columns)  # False → it's just a property
TL;DR:

If it's a...	Use ()?	Example
Function or method	✅ Yes	np.sum(), df.head()
Attribute or property	❌ No	df.columns, df.shape

# 🔹 loc[ ] vs. iloc[]: What’s the difference?
✅ loc[ ] → Selects data using labels (column names & row index labels)
.
✅ iloc[ ] → Selects data using integer index position

🔍 Output Insight:
 ✔ df.loc['A'] selects the row where the index is 'A'.
 ✔ df.iloc[1] selects the second row (since Python indexing starts at 0).
 ✔ loc[] is best when working with named indices, while iloc[ ] is best for numerical indexing.

🎯 Key Takeaways:
✔ Use loc[ ] for label-based selection.
✔ Use iloc[ ] for position-based selection.
✔ Combine them with conditions for advanced filtering (e.g., df.loc[df['Age'] > 30]).

Pandas .set_index()  Method – Beginner’s Guide
The pandas DataFrame.set_index()  method reassigns the row index of a DataFrame using one or
more existing columns (or arrays) as the new labels. By default, pandas uses simple integer indices (0, 1, 2,
…), but you can use set_index()  to make a meaningful column (like dates, names, etc.) the index. For
example,  df.set_index('Date')  would make the “Date” column the row labels, so you can look up
rows by date. As the pandas docs explain: “Set the DataFrame index (row labels) using one or more existing
columns or arrays.”
. This can simplify data lookup and operations (e.g. selecting rows by label) and is
often used when preparing data. 
Syntax and Parameters
The method signature is: 
DataFrame.set_index(keys, *, drop=True, append=False, inplace=False, verify_integrity=False)
Key parameters include:
keys : Label or list of labels/arrays. Specifies which column(s) (or arrays) to use as the new index. You
can pass a single column name (string) or a list of names for multiple columns. You can also pass an
array or pandas Series of the same length as the DataFrame. (The docs note: it “can be a single
column key, a single array of the same length, or a list containing a combination of column keys and
arrays.”)
.
drop  (bool, default True ): If True , the original column(s) used for the index are removed from
the DataFrame’s columns. This is the usual behavior because once a column becomes the index, it’s
no longer a regular column. If you want to keep the column in the data, set drop=False
.
For  example,  df.set_index('col',  drop=False)  will  keep  “col”  as  both  the  index  and  a
column.  By  default,  drop=True  “causes  set_index  to  drop  the  column  when  it  creates  the
index”
.
append  (bool, default False ): If True , the new column(s) are appended to the existing index,
creating a MultiIndex. If False , the existing index is replaced. (Using multiple keys usually replaces
the index with a new MultiIndex, which is like having append=False  on multiple keys.) The default
False  means the old index is discarded (unless you explicitly append).
inplace  (bool, default False ): If False , set_index()  returns a new DataFrame with the
new index, leaving the original unchanged. If True , it modifies the original DataFrame in place
and returns None

 By default, pandas leaves the original intact and creates a copy with the
new  index.  (The  pandas  docs  note  that  the  return  value  is  “DataFrame  or  None  if
inplace=True ”) Setting inplace=True  means you won’t need to assign the result to a new 
variable,  but  it  also  means  the  method  returns  nothing.  For  example,  
df.set_index('A',inplace=True)  will change df  directly.

verify_integrity  (bool, default False ): If True , pandas checks whether the new index has
duplicate labels, and will raise an error if duplicates are found. By default this check is skipped (to be
faster), allowing duplicate index values. If you need a unique index, set verify_integrity=True .
As one guide notes,  “by default, if there are duplicates in the new index, they are used as is. Setting
verify_integrity=True  results in an error if duplicates exist.”
. Use this to catch unintended duplicate index values.
In summary, the simplest call is df.set_index('col')  (using one column). For multiple columns, use a
list like  df.set_index(['col1','col2'])  (which creates a MultiIndex). The key parameters let you
control whether to drop columns, append to an existing index, modify in place, or enforce unique labels
.

Tips & Common Mistakes
Forgetting to assign or use inplace : By default, set_index()  returns a new DataFrame. If you
just  call  df.set_index('col')  and  do  not  assign  it  or  use  inplace=True ,  your  original
DataFrame  df  will  not  change.  Always  assign  it  ( df  =  df.set_index(...) )  or  use
inplace=True  if you intend to keep the changes
. 
Dropping the column: Remember that by default set_index()  removes the column you use as
the index. If you need to keep that data, use  drop=False
. A common surprise is doing
df.set_index('col')  and then later wondering where the col  data went – it was removed
from the DataFrame. 
Duplicate index values: If the column you use for the index has duplicate values, pandas will allow
them  by  default  (so  you’ll  have  a  non-unique  index).  If  you  really  need  a  unique  index,  set
verify_integrity=True  and pandas will raise an error when duplicates exist
. It’s a good
idea  to  be  aware  of  this  if  you  rely  on  unique  indexing.  For  example,  using
verify_integrity=True  will catch mistakes early.
Using inplace=True  carefully: Setting inplace=True  modifies the original object and returns
None . This can save memory if you do not need a copy, but pandas notes that “inplace” operations
often do not give a performance boost and are sometimes discouraged. Make sure to use it only
when you want to change the DataFrame without assigning a new variable
.
MultiIndex ordering: When setting multiple columns as index, the order you list them matters – it
defines  the  hierarchy  of  the  MultiIndex.  For  example,
 ['year','month']  
vs
['month','year']  will produce different row-label orderings as shown above. Choose the order
that makes sense for your data.
Appending to an existing index: The append=True  option can add new index levels to an existing
index. This is less common, but useful if you want to keep the old index and stack new labels on top.
Resetting index to recover columns: If you accidentally overwrite your data (e.g. by setting index
inplace and dropping the column), you can often use reset_index()  to bring the index back as a
column
.  For  example,  df.reset_index()  will  move  the  current  index  back  into  the
columns and restore a default integer index.
By following these tips and understanding the parameters, you can avoid common pitfalls (like accidentally
dropping needed data or losing track of what your index is). Use  set_index()  whenever a column
naturally identifies rows (dates, IDs, categories, etc.), and remember you can always switch back with
reset_index()
