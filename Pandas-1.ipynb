````python
g3 = df.groupby('pickup_borough')
print(type(g3))
g3.groups

'''
Why groups useful?

To know which rows belong to each category
To manually access grouped rows
For debugging and understanding groupby
For custom operations on specific groups 

groupby() creates groups
.groups shows the row-index mapping of those groups
'''
````

Because `g3`, `g2`, and `g1` are **different types of objects**.

### 1. `g3`

```python id="g1a2bc"
g3 = df.groupby('pickup_borough')
type(g3)
```

`g3` is a **GroupBy object**.

Type:

```python id="h3d4ef"
pandas.core.groupby.generic.DataFrameGroupBy
```

A `GroupBy` object stores:

* the groups
* indexes
* grouping logic

So it has attributes like:

```python id="j5k6lm"
g3.groups
```

---

### 2. `g2`

```python id="n7p8qr"
g2 = df.groupby('pickup_borough')[['pickup_borough']].count()
```

Here, `.count()` is already executed.

So `g2` becomes a normal **DataFrame**.

Type:

```python id="s1t2uv"
pandas.core.frame.DataFrame
```

DataFrames do NOT have `.groups`.

---

### 3. `g1`

```python id="w3x4yz"
g1 = df.groupby('pickup_borough').size()
```

`.size()` returns counts directly.

So `g1` becomes a **Series**.

Type:

```python id="a5b6cd"
pandas.core.series.Series
```

Series also do NOT have `.groups`.

---

### Simple rule

| Variable | Type           | Has `.groups`? |
| -------- | -------------- | -------------- |
| `g3`     | GroupBy object | ✅ Yes          |
| `g2`     | DataFrame      | ❌ No           |
| `g1`     | Series         | ❌ No           |

`.groups` exists only before aggregation (`count()`, `size()`, `sum()`, etc.) is applied.

````python
g3.get_group('Manhattan').head()
# get_group() returns all rows belonging to a specific group.

'''.groups → gives row indexes
.get_group() → gives actual rows/data of that group'''

print(g3.groups.keys())
# .groups.keys() lists all unique group names (categories) in the grouping column.

````

````python

print(g3.groups['Manhattan'][:5])
# .groups['Manhattan'] gives row indexes for the 'Manhattan' group. [:5] shows the first 5 indexes.


````

````python
print(g3.get_group('Manhattan')[['pickup_borough']].head())
# get_group('Manhattan') returns all rows for 'Manhattan'. [['pickup_borough']] selects just that column. head() shows the first 5 rows of that selection.
````

````python
g3.ngroups
# .ngroups gives the total number of unique groups formed. In this case, it counts how many unique pickup boroughs there are in the dataset.
print(g3.ngroups)
````

````python
g3.size()
````

````python
for name, data in g3:
    print(name)
    print(data.head(2))
    print('---')
# Looping through groups:
# name → the group name (e.g., 'Manhattan')
# data → the DataFrame containing rows of that group
# head(2) shows the first 2 rows of each group for demonstration.
````

A `GroupBy` object can be iterated in different ways.

Example:

```python id="m4q8tw"
g3 = df.groupby('pickup_borough')
```

---

### 1. Iterate group name + data (most common)

```python id="k7n2vx"
for name, data in g3:
    print(name)
    print(data)
```

* `name` → group name
* `data` → DataFrame of that group

---

### 2. Iterate only group names

```python id="r5p1lz"
for name in g3.groups:
    print(name)
```

Output:

```python id="x8d4qm"
Manhattan
Queens
Brooklyn
```

---

### 3. Iterate group name + row indexes

```python id="c2v7wy"
for name, indexes in g3.groups.items():
    print(name, indexes)
```

Output:

```python id="h6m3pt"
Manhattan Int64Index([0,2], dtype='int64')
Queens Int64Index([1], dtype='int64')
```

---

### 4. Iterate using `indices`

```python id="u9k5rx"
for name, indexes in g3.indices.items():
    print(name, indexes)
```

Similar to `.groups`.

---

### 5. Iterate aggregated results

```python id="a1t8mn"
for name, value in g3['fare'].sum().items():
    print(name, value)
```

Output:

```python id="e4q7zs"
Manhattan 30
Queens 15
```

---

### Important understanding

`GroupBy` iteration gives:

```python id="d5w2lp"
(group_name, grouped_dataframe)
```

# .items() is used to iterate through key-value pairs.
'''
| Method      | Returns           |
| ----------- | ----------------- |
| `.keys()`   | only keys         |
| `.values()` | only values       |
| `.items()`  | key + value pairs |
'''

So internally:

```python id="f8x6kr"
for x in g3:
```

each `x` looks like:

```python id="p3n9tv"
('Manhattan', <DataFrame>)
```
