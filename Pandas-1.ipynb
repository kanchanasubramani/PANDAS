```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('taxis')
```

```python
df.head()
```

```python
df[['pickup_borough']].unique()
```

```python
unique_1=df[['pickup_borough']].drop_duplicates()
```

```python
unique_1
```

```python
unique_2=df['pickup_borough'].unique()
unique_2
```

* `df['pickup_borough'].unique()` → returns unique values as a **NumPy array**. Fast and simple.
* `df[['pickup_borough']].drop_duplicates()` → returns unique rows as a **DataFrame**, preserving index and column structure.

Use:

* `.unique()` for quick value lists
* `.drop_duplicates()` for DataFrame/table operations
