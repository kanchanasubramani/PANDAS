Why pd.set_option('display.max_rows',None) not ecom.set_option('display.max_rows',None):

 pd.set_option() is part of pandas, not the DataFrame.
set_option() is a global configuration function from the pandas library.

It changes the way pandas displays output in general, not just for one DataFrame.
That affects how all DataFrames display in your environment — telling pandas, "Hey, stop truncating rows when printing. Show me everything."

ecom is just a pandas DataFrame object.

A DataFrame doesn’t have any method called set_option().

So ecom.set_option() would throw an AttributeError, like:
AttributeError: 'DataFrame' object has no attribute 'set_option'

🧠 Analogy
Think of pd.set_option() like changing the settings of a TV (resolution, brightness, etc.) — it affects everything you watch on it.

Meanwhile, ecom is like just one show you're watching. It doesn’t control the TV’s settings.

# 🔹 loc[ ] vs. iloc[]: What’s the difference?
✅ loc[ ] → Selects data using labels (column names & row index labels)
.
✅ iloc[ ] → Selects data using integer index position

🔍 Output Insight:
 ✔ df.loc['A'] selects the row where the index is 'A'.
 ✔ df.iloc[1] selects the second row (since Python indexing starts at 0).
 ✔ loc[] is best when working with named indices, while iloc[ ] is best for numerical indexing.

🎯 Key Takeaways:
✔ Use loc[ ] for label-based selection.
✔ Use iloc[ ] for position-based selection.
✔ Combine them with conditions for advanced filtering (e.g., df.loc[df['Age'] > 30]).
